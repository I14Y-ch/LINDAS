```python
from requests import Session
from requests.auth import HTTPBasicAuth
```

```python
# Authentication and proxies 

# auth = HTTPBasicAuth('', '') #TEST
#auth = HTTPBasicAuth('', '') #INT
auth = HTTPBasicAuth('', '') #PROD


proxies = {
          "http":'http://proxy-bvcol.admin.ch:8080',
          "https":'http://proxy-bvcol.admin.ch:8080'
        }
```

```python
#######################
# upload Turtle RDF data to the graph 'https://lindas.admin.ch/fso/i14y'
#######################
# Ã¼berschreiben https://docs.stardog.com/operating-stardog/database-administration/adding-data#add-data-using-the-graph-store-http-protocol 

headers_file_i14y_to = {
                        'Content-Type': 'text/turtle; charset=UTF-8'                      
                        }

params = {'graph': 'https://lindas.admin.ch/fso/i14y'} #optional
data=open('file.ttl','r', encoding='utf-8').read()

session = Session()
session.trust_env = False  


r2 = session.put(url='https://stardog-test.cluster.ldbar.ch/lindas', #test
                  params=params,
                  headers=headers_file_i14y_to, 
                  proxies = proxies,
                  auth=auth,
                  data=data.encode('utf-8'), 
                  verify=False)
print(f"Response Status Code: {r2.status_code}")

print(f"Response Headers: {r2.headers}")
print(f"Response Content: {r2.text}")
```

```python
#######################
# retrieve data with SPARQL queries 
#######################

headers_sparql_query = {
    'Accept': 'application/sparql-results+json',  
}


sparql_query = """
SELECT ?subject ?predicate
WHERE {
  ?subject ?predicate <https://ld.admin.ch/cube/dimension/hierarchy/yes-no> .
}

"""


sparql_endpoint = 'https://stardog-test.cluster.ldbar.ch/lindas/query'
params = {'graph': 'https://lindas.admin.ch/fso/i14y', 'query': sparql_query}
session = Session()
session.trust_env = False 


response = session.get(
    url=sparql_endpoint,
    headers=headers_sparql_query,
    params=params,
    proxies=proxies,
    auth=auth,
    verify=False 
)


print(f"Response Status Code: {response.status_code}")
print(f"Response Headers: {response.headers}")
print(f"Response Content: {response.text}")
```

```python
#######################
# DELETE data
#######################


headers_sparql_update = {
    'Content-Type': 'application/sparql-update',  
    'Accept': 'application/sparql-results+json'  
}

#query to delete all the data in the i14y graph
sparql_delete_query = """
DELETE WHERE {
    GRAPH <https://lindas.admin.ch/fso/i14y> {   
        ?s ?p ?o .
    }
}
"""


sparql_endpoint = 'https://stardog-test.cluster.ldbar.ch/lindas/update'

session = Session()
session.trust_env = False  

response = session.post(
    url=sparql_endpoint,
    headers=headers_sparql_update,
    data=sparql_delete_query,  # The DELETE query as the body
    proxies=proxies,
    auth=auth,
    verify=False  
)


print(f"Response Status Code: {response.status_code}")
print(f"Response Headers: {response.headers}")
print(f"Response Content: {response.text}")
```
